Linear Regression


```python
import numpy as np
x = np.array([1, 2, 3, 4, 5]) # hours studied
y = np.array([40, 50, 60, 70, 80]) # marks obtained
n = len(x) # number of students

# Step 1 : calculate sums
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x ** 2)

# Step 2 : Calculate slope(m) ans intercept(c)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
c = (sum_y - m * sum_x) / n

# Step 3 : Predict the marks using the line y = mx + c
y_pred = m*x + c

# Step 4 : Show result
print(f"Slope(m) : {m : .2f}")
print(f"Intercept(c) : {c : .2f}")
print("Predicted marks :", np.round(y_pred).astype(int))

# Step 5 : Draw picture
import matplotlib.pyplot as plt
plt.scatter(x, y, color = 'blue', label = 'Students')
plt.plot(x, y_pred, color = 'red', label = 'Regression line')
plt.xlabel('Hours studied')
plt.ylabel('Marks obtained')
plt.title('Hours studied vs Marks obtained')
plt.grid(True)
plt.legend()
plt.show()
```
