```python
#2. Study of python basic libraries
import math

num = 9
angle_in_degrees = 90

print("Square root:", math.sqrt(num))
print("Factorial:", math.factorial(5))
print("Sine of 90 degrees:", math.sin(math.radians(angle_in_degrees)))
print("Logarithm of base 10:", math.log10(100))
print("Value of Pi:", round(math.pi, 2))
```

```python
#Usage of ceil function in math library
import math

num1 = 4.3
num2 = 9.7
num3 = -2.8
num4 = 5.0

print("Ceil of", num1, "is: ", math.ceil(num1))
print("Ceil of", num2, "is: ", math.ceil(num2))
print("Ceil of", num3, "is: ", math.ceil(num3))
print("Ceil of", num4, "is: ", math.ceil(num4))
```
